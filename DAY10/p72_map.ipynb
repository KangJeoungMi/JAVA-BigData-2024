- 지도 타일 셋 지정
    - OpenStreetMap(기본), OpenStreetMap.HOT, OpenStreetMap.BZH(글자 없이 지도만 나옴), OpenTopoMap(로딩 오래 걸림)
    - CartoDB Positron, CartoDB PositronNoLabels, CartoDB Voyager, CartoDB Dark Matter(CartoDB 그룹)
    - CyclOSM

````python
#타일 셋 지정
map = folium.Map(location=(35.178508, 129.199443), tiles='CartoDB Voyager', zoom_start=17)
map
````

- 지도마커 표시
    - Marker() 클래스로 생성한 뒤 지도에 추가

````python
# 마커
folium.Map(location=(35.178508, 129.199443), tiles='CartoDB Voyager', zoom_start=17)
folium.Marker(location=[35.178508, 129.199443], tooltip='송정해수욕장', popup='서핑').add_to(map)
popup=folium.Popup('<pre><a href="">스타벅스</a>로 오세요</pre>', min_width=200,max_width = 300)
folium.Marker(location=[35.180762, 129.202324], tooltip='스타벅스 부산송정비치점', popup=popup,icon=folium.Icon(color='green', icon='heart')).add_to(map)

map
````
